```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# 1. Set the visual style
sns.set_theme(style="whitegrid")
# 2. Load the dataset
df = sns.load_dataset('tips')
# 3. Preview the data
print("Successfully loaded! Here are the first 5 rows:")
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='total_bill', bins=20, kde=True, color='teal')
plt.title('Distribution of Total Restaurant Bills')
plt.xlabel('Bill Amount ($)')
plt.ylabel('Number of Tables')
plt.show()
```

```python
# The Scatter Plot (Relationships)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='total_bill', y='tip', hue='smoker', size='size', alpha=0.7)
plt.title('Relationship: Bill Size vs. Tip Amount')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.legend(title='Smoker Status')
plt.show()
```

```python
# The Box Plot (Comparison)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='day', y='total_bill', hue='day', palette='Set2', legend=False)
plt.title('Spending Habits by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.show()
```
