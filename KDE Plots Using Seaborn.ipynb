```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
```

```python
#Part 3: Univariate KDE Plot
sns.kdeplot(data=df, x='total_bill', fill=True)
plt.title('Univariate KDE: Total Bill')
plt.show()
```

```python
#Part 4: Grouped Univariate KDE Plot
sns.kdeplot(data=df, x='total_bill', hue='sex', fill=True)
plt.title('Grouped Univariate KDE: Total Bill by Sex')
plt.show()
```

```python
#Part 5: Bivariate KDE Plot
sns.kdeplot(data=df, x='total_bill', y='tip', fill=True)
plt.title('Bivariate KDE: Total Bill vs Tip')
plt.show()
```
