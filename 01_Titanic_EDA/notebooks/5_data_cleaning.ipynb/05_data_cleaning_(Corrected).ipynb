```python
# Notebook 5: Data Cleaning (Warning-Free Version)

import pandas as pd

# --------------------------
# Step 1: Load Raw Data
# --------------------------
train_path = r"C:\Users\HP\Documents\GitHub\DataMining-Portfolio\01_Titanic_EDA\Data\raw\train.csv"
test_path  = r"C:\Users\HP\Documents\GitHub\DataMining-Portfolio\01_Titanic_EDA\Data\raw\test.csv"

train = pd.read_csv(train_path)
test  = pd.read_csv(test_path)

# --------------------------
# Step 2: Fill Missing Values Safely
# --------------------------

# Fill missing Age with median
train['Age'] = train['Age'].fillna(train['Age'].median())
test['Age']  = test['Age'].fillna(test['Age'].median())

# Fill missing Embarked with mode
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])

# --------------------------
# Step 3: Drop Columns with Too Many Missing Values
# --------------------------
train = train.drop(columns=['Cabin'])
test  = test.drop(columns=['Cabin'])

# --------------------------
# Step 4: Save Cleaned Data
# --------------------------
train.to_csv(r"C:\Users\HP\Documents\GitHub\DataMining-Portfolio\01_Titanic_EDA\Data\processed\train_clean.csv", index=False)
test.to_csv(r"C:\Users\HP\Documents\GitHub\DataMining-Portfolio\01_Titanic_EDA\Data\processed\test_clean.csv", index=False)

# --------------------------
# Step 5: Verify Cleaning
# --------------------------
print("Missing values in train dataset:\n", train.isnull().sum())
print("\nMissing values in test dataset:\n", test.isnull().sum())
```
